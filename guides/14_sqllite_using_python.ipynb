## Module 1: What is a Database? (The Absolute Basics)
Before writing code, let us understand what a database actually is.
## The Grocery List Analogy
Imagine you are writing down a grocery list.

* You could write it on a scrap of paper (Text File).
* You could organize it into rows and columns using Microsoft Excel or Google Sheets (Spreadsheet).

A spreadsheet works great for a small list. But what if you run a massive supermarket chain like Walmart? You have millions of products, thousands of employees, and millions of customers buying items every second.

* A spreadsheet would crash.
* Multiple people cannot update it at the exact same time safely.
* Finding one specific receipt from three years ago would take forever.

This is why we use a Database. A database is a highly organized, digital filing cabinet designed to store, retrieve, and manage massive amounts of data instantly and safely. 
## What is SQL and SQLite?

* SQL (Structured Query Language): This is the universal language we use to "talk" to databases. Pronounced like "Sequel," it allows us to ask the database questions like, "Show me all users who bought an item today."
* SQLite: This is a specific database engine. Unlike big databases (like MySQL or PostgreSQL) that require complex software installations and servers, SQLite is built right into Python. It stores your entire database in a single, simple file on your computer. It is lightweight, fast, and perfect for learning, prototyping, and mobile apps.

## Module 2: Setting Up and Making Your First Connection
To use SQLite in Python, you do not need to install anything using pip. Python comes with a built-in library called sqlite3. 
## Step 1: Import the Library
We tell Python we want to use the SQLite library. 
## Step 2: Connect to a Database
We use sqlite3.connect('filename.db'). If the file does not exist, Python will automatically create it for you!  
## Step 3: Create a Cursor  
Think of the Cursor as a digital blinking pointer or a robotic hand. The connection opens the door to the database, but the cursor is what actually executes your commands and grabs the results. 
## Step 4: Close the Connection
Always close your database when you are done. If you leave it open, the file can get locked or corrupted.


```python
# Cell 1: Creating your very first database file
import sqlite3
# 1. Connect to the database (creates 'school.db' if it doesn't exist)
connection = sqlite3.connect('school.db')
print("Successfully connected to the database!")
# 2. Create a cursor object to execute commands
cursor = connection.cursor()
# 3. Always close the connection when finished
connection.close()
print("Database connection closed safely.")
```


## Module 3: Creating Your First Table (DDL)
Databases organize data into Tables. Think of a table exactly like an Excel sheet. It has rows (individual records) and columns (the properties of that record). 
Before adding data, we must define the structure of the table. This is called the Schema. We must tell the database exactly what columns we want and what type of data belongs in them. 
## SQLite Data Types:

   1. TEXT: For words, names, and sentences.
   2. INTEGER: For whole numbers (like age, count, IDs).
   3. REAL: For decimal numbers (like prices, GPA, height).
   4. NULL: For missing or empty data. [29, 30, 31, 32, 33] 

## Key Concepts:

* PRIMARY KEY: A unique identifier for every row. Think of it like your Aadhaar card number or Student ID. No two rows can ever have the same Primary Key.
* AUTOINCREMENT: Tells the database to automatically number the rows for us (1, 2, 3, etc.).


```python
# Cell 2: Creating a Table
import sqlite3
# Connect and get a cursor
conn = sqlite3.connect('school.db')
cursor = conn.cursor()
# Define the SQL command to create a table# We use triple quotes (''' ) to write multi-line strings easily
create_table_query = '''
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER,
    gpa REAL
);
'''
# Execute the command
cursor.execute(create_table_query)
print("Table 'students' created successfully!")
# Commit (save) changes and close
conn.commit()
conn.close()
```



Note: IF NOT EXISTS is a safety check. It prevents Python from crashing if you run this cell a second time.
## Module 4: Inserting and Saving Data (The Crucial "Commit")
Now that our table structure exists, let us add some rows of data.
## The Concept of a "Transaction" and "Commit"
When you add, delete, or change data in a database, the changes are stored in a temporary "holding zone." The database is waiting to see if you make a mistake.

* To permanently save the changes to the hard drive, you must call connection.commit().
* If you do not commit, your changes will disappear the moment you close the program! 

```python
# Cell 3: Inserting data into our table
import sqlite3
conn = sqlite3.connect('school.db')
cursor = conn.cursor()
# SQL query to insert a single row
insert_query = '''
INSERT INTO students (name, age, gpa) 
VALUES ('Rahul Sharma', 20, 8.5);'''
# Execute the insert
cursor.execute(insert_query)
# CRITICAL: Save the changes!
conn.commit()
print("Successfully inserted 1 student record!")

conn.close()
```


## Module 5: Reading and Fetching Data (Queries)
To get data out of a database, we use the SELECT statement.

* SELECT * FROM students means "Grab every single column from the students table."

After running cursor.execute(), the data is sitting inside the cursor. We use fetch methods to pull it into Python variables: 

   1. fetchall(): Grabs every row matching the query as a list of tuples.
   2. fetchone(): Grabs just the very first row. 

# Cell 4: Fetching data from the database

```python


import sqlite3
conn = sqlite3.connect('school.db')
cursor = conn.cursor()
# Run the query
cursor.execute("SELECT * FROM students;")
# Fetch all records
all_students = cursor.fetchall()

print("Data retrieved from database:")
print(all_students) 
# Output will look like a list of tuples: [(1, 'Rahul Sharma', 20, 8.5)]

print("\nFormatting the output nicely:")
for student in all_students:
    print(f"ID: {student[0]} | Name: {student[1]} | Age: {student[2]} | GPA: {student[3]}")

conn.close()
```

## Module 6: Intermediate Concepts (Parameters & Bulk Operations)
As a programmer, you should almost never hardcode values into your SQL strings.
## The Danger: SQL Injection
If you use Python f-strings like f"INSERT INTO students VALUES ('{user_input}')", a malicious user can type SQL code into your input box and erase your entire database. This is a massive security flaw called SQL Injection.
## The Solution: Parameterized Queries
We use a question mark ? as a placeholder. We then pass the actual data as a tuple inside the .execute() method. SQLite safely sanitizes the data for us. 
## Bulk Insertion with executemany
If you have a list containing thousands of rows, executing them one by one in a loop is incredibly slow. Instead, use cursor.executemany(). It sends all the data to the database in one highly optimized batch.  

# Cell 5: Safe parameterized insertions and bulk loading

```python
import sqlite3
conn = sqlite3.connect('school.db')
cursor = conn.cursor()
# 1. Safe Single Insertion using placeholders (?)
safe_query = "INSERT INTO students (name, age, gpa) VALUES (?, ?, ?);"
single_data = ("Priya Patel", 21, 9.2)

cursor.execute(safe_query, single_data)
# 2. Bulk Insertion using executemany
many_students = [
    ("Amit Mishra", 19, 7.8),
    ("Sneha Reddy", 22, 9.5),
    ("Vikram Singh", 20, 6.4)
]

cursor.executemany(safe_query, many_students)
# Save everything
conn.commit()
print("Successfully inserted multiple records safely!")

conn.close()
```

## Module 7: Updating and Deleting Data
Data is dynamic. Students change their names, improve their grades, or leave the school. We manage this using UPDATE and DELETE. 
## The Golden Rule: Always Use a WHERE Clause
If you execute UPDATE students SET gpa = 10.0; without a WHERE clause, every single student in your database will have their GPA changed to 10.0. Always specify which rows you want to affect!

# Cell 6: Updating and Deleting records

```python
import sqlite3
conn = sqlite3.connect('school.db')
cursor = conn.cursor()
# 1. Update Amit's GPA (Student ID 3)
update_query = "UPDATE students SET gpa = ? WHERE student_id = ?;"
cursor.execute(update_query, (8.2, 3))
# 2. Delete Vikram from the database (Student ID 5)
delete_query = "DELETE FROM students WHERE student_id = ?;"
cursor.execute(delete_query, (5,)) # Note: (5,) is a one-item tuple in Python

conn.commit()
print("Update and Delete operations completed successfully.")
# Let's see the clean final table
cursor.execute("SELECT * FROM students;")
print(cursor.fetchall())

conn.close()
```

## Module 8: Advanced Pythonic Patterns (Context Managers)
As you build advanced applications, manually opening, committing, and closing connections becomes tedious and error-prone. If your code crashes midway, the connection stays open forever.
## The with Statement
Python provides context managers (with blocks). When you use a context manager with SQLite:  

* It automatically commits your changes if the code runs successfully.
* It automatically rolls back (cancels) your changes if your Python code encounters an error. 

# Cell 7: Professional Clean Code using Context Managers

```python

import sqlite3
# The 'with' block manages the connection automatically
with sqlite3.connect('school.db') as conn:
    cursor = conn.cursor()
    
    # Try inserting a new record
    cursor.execute("INSERT INTO students (name, age, gpa) VALUES (?, ?, ?);", ("Ananya Rao", 23, 8.9))
    
    # Notice: No conn.commit() is written here! 
    # Python automatically commits when the 'with' block finishes successfully.

print("Database operations completed and saved automatically!")
```

## Module 9: Advanced Database Features (Error Handling & Row Factories)
Let us bridge the gap to professional production-grade code. We will implement two advanced concepts:

   1. Error Handling (try-except): Prevents your app from crashing if a database error happens (e.g., database file is corrupted, or table name misspelled).
   2. Row Factory: By default, SQLite returns data as plain tuples (1, 'Rahul', ...). You cannot access items by their column names. Setting conn.row_factory = sqlite3.Row allows us to interact with rows like Python dictionaries! 

# Cell 8: Advanced Error Handling and Dictionary-like Access

```python
import sqlite3
try:
    # Use context manager
    with sqlite3.connect('school.db') as conn:
        
        # MAGIC LINE: Change how rows are returned
        conn.row_factory = sqlite3.Row
        
        cursor = conn.cursor()
        
        # Intentional typo in query to test error handling (e.g., student_id_typo)
        cursor.execute("SELECT student_id, name, gpa FROM students;")
        
        rows = cursor.fetchall()
        
        print("Advanced Data Fetching Results:")
        for row in rows:
            # Now we can read data using column names instead of index numbers!
            print(f"Student: {row['name']} has a GPA of {row['gpa']}")
except sqlite3.Error as error:
    print(f"An error occurred while working with SQLite: {error}")

```
