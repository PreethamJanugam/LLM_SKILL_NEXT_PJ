**Tools and Tool Calls**.

Think of a standard LLM like a brilliant student who has read millions of books but is locked in a room without internet, a calculator, or a watch. They know a lot of theory, but they can't check today's weather or solve a massive math problem instantly.

**Tools** are how we let that student interact with the outside world.

---

## 🛠️ What is a "Tool" in LLMs?

A **Tool** (also known as *Function Calling*) is an external capability given to an LLM so it can look up real-time information, perform complex calculations, or take actions.

Instead of guessing an answer, the LLM can say, *"Hey, I don't know this, let me use a tool to find out!"*

### Common Examples of Tools:

* **A Web Search Tool:** To find today's news.
* **A Calculator Tool:** To solve precise math equations ($12345 \times 67890$) without making a typo.
* **A Database Tool:** To check if a product is in stock for an e-commerce chatbot.

---

## 🔄 The Step-by-Step Process of a Tool Call

Let’s look at exactly what happens behind the scenes when an LLM uses a tool. We will use a simple example: **The user asks, "What is the weather in Tokyo right now?"**

### Step 1: The Setup (Providing the Toolbox)

Before the user even types a question, the developer gives the LLM a list of available tools. This list is written in code (usually JSON) and describes exactly what the tool does and what information it needs.

> **Example description given to LLM:** `"You have a tool called 'get_weather'. It requires a 'location' string (e.g., 'Tokyo')."`

### Step 2: The User Asks a Question

The user submits their prompt: *"What is the weather in Tokyo?"*

### Step 3: The LLM Decides and Decides to "Call"

The LLM realizes it doesn't know the live weather (since its training data is in the past). It looks at its toolbox, sees the `get_weather` tool, and decides to use it.
Instead of answering the user, the LLM outputs a **Tool Call**. It essentially shoots back a piece of code saying:

> `"I want to use 'get_weather' for location='Tokyo'."`

### Step 4: The Application Executes the Tool

The LLM itself cannot actually browse the live internet. Instead, the computer program running the LLM catches that "Tool Call" request, goes out to a real weather website, and grabs the live data.

> **Result found by the system:** `"It is currently 18°C and rainy in Tokyo."`

### Step 5: Feeding the Answer Back

The system sends this live result *back* to the LLM.

### Step 6: The Final Answer

Now armed with the correct, real-time facts, the LLM generates a nice, conversational response for the user.

> **LLM Response to user:** `"Right now in Tokyo, it is 18°C and raining, so you might want to grab an umbrella if you're heading out!"`

---

## 📊 Summary of the Workflow

Here is how the data flows in a neat, quick-reference table:

| Step | Who is acting? | What are they doing? |
| --- | --- | --- |
| **1. Request** | User | Asks a question requiring real-time/exact data. |
| **2. Decision** | LLM | Recognizes it needs help and outputs a **Tool Call** structured request. |
| **3. Execution** | External System | Runs the actual code/API to get the real data. |
| **4. Response** | LLM | Receives the data and writes a natural response back to the user. |

---

## 💡 Why is this a Game-Changer?

1. **Reduces Hallucinations:** LLMs are notorious for confidently making up facts when they don't know the answer. Tools force them to rely on hard data.
2. **Enables Action:** With tools, an LLM doesn't just *talk*; it can *do*. It can book a flight, send an email, or update a row in a spreadsheet if given the right tools.


### Step 1: The Developer Defines the Tool (The Blueprint)

First, you write a standard Python function that talks to your database. Then, you describe this function to the LLM using a structured format (usually JSON) so the LLM knows it exists.

```python
def get_student_gpa(student_id: str) -> str:
    # In real life, this would query a secure university database
    database = {"S12345": "3.85", "S98765": "3.20"}
    return database.get(student_id, "Student ID not found.")

# 2. This is the "blueprint" description we pass to the LLM
gpa_tool_blueprint = {
    "name": "get_student_gpa",
    "description": "Looks up a student's current GPA using their unique Student ID.",
    "parameters": {
        "type": "object",
        "properties": {
            "student_id": {
                "type": "string",
                "description": "The 6-character student ID starting with 'S' (e.g., S12345)"
            }
        },
        "required": ["student_id"]
    }
}
```


### Step 2: The User Triggers the Conversation

The user types:

> *"Can you check my current GPA? My student ID is S12345."*

---

### Step 3: The LLM Generates a "Tool Call"

The LLM reads the user's prompt and scans its available blueprints. It realizes it cannot guess a GPA, but it sees the `get_student_gpa` tool.

Instead of writing a text sentence back to the user, the LLM outputs a structured **Tool Call** request:
#### it's in json format

```python

{
  "tool_call": {
    "name": "get_student_gpa",
    "arguments": {
      "student_id": "S12345"
    }
  }
}
```


### Step 4: The Application Executes the Code

Your Python application intercepts that JSON message from the LLM, reads the arguments, and actually runs the Python function we wrote in Step 1.


```python
# The system automatically runs:
tool_result = get_student_gpa(student_id="S12345")

print(tool_result) 
# Output from database: "3.85"
```



### Step 5: The Final Answer

The application sends the result (`"3.85"`) back to the LLM. The LLM merges this hard fact with its natural language skills to give the student a polite response.

> **LLM Response:** *"Your current GPA is 3.85. Great job keeping your grades up!"*



### 🔍 Key Takeaway for a Fresher

Notice the clear division of labor in the diagram workflow:

* **The LLM** is the **brain**. It doesn't actually touch the database; it just *decides* which tool to use and extracts the right argument (`S12345`).
* **The Python Code** is the **muscle**. It takes the LLM's decision, does the heavy lifting (running the database query), and hands the answer back.
