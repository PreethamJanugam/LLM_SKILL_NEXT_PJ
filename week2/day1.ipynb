## Finally - my personal fave - the wonderfully lightweight LiteLLM

## Core Differences
LangChain and LiteLLM serve fundamentally different purposes in the LLM ecosystem. LangChain is a heavy framework for building complex, multi-step applications. LiteLLM is a lightweight translator that unifies different model APIs into a single format.  
## Key Comparison

* Primary Purpose: LangChain builds applications. LiteLLM simplifies API calls.
* Architecture: LangChain uses complex abstractions like chains and agents. LiteLLM uses a single, standardized interface.
* API Structure: LangChain introduces unique syntax. LiteLLM replicates the standard OpenAI format for all providers.
* Supported Models: LangChain supports extensive custom components. LiteLLM translates over 100 specific LLM APIs.
* Overhead: LangChain has a steep learning curve. LiteLLM requires minimal setup. 

## Detailed Feature Breakdown## 1. Implementation Scope

* LangChain: Best for creating autonomous agents, retrieval-augmented generation (RAG) pipelines, and memory-infused chatbots.
* LiteLLM: Best for teams switching between model providers (like Anthropic, OpenAI, and Cohere) without rewriting codebases. 
## 2. Model I/O and Formatting

* LangChain: Packages prompts into specialized wrappers. It uses unique parsers to handle model outputs.
* LiteLLM: Passes standard dictionaries directly to the model. It returns data structured exactly like an OpenAI completion response. 

## 3. State and Memory Management

* LangChain: Features built-in memory utilities to track conversation history automatically across multiple turns.
* LiteLLM: Stateless by design. It requires developers to manage chat history manually in their application logic.  

## 4. Ecosystem Integration

* LangChain: Connects natively with hundreds of vector databases, data loaders, and external tools.
* LiteLLM: Focuses narrowly on model hosting, load balancing, fallback routing, and cost tracking.

## And now for some fun - an adversarial conversation between Chatbots..

You're already familar with prompts being organized into lists like:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "user prompt here"}
]
```

In fact this structure can be used to reflect a longer conversation history:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "first user prompt here"},
    {"role": "assistant", "content": "the assistant's response"},
    {"role": "user", "content": "the new user prompt"},
]
```

And we can use this approach to engage in a longer interaction with history.

````python
# Let's make a conversation between gemma and llama
# We're using cheap versions of models so the costs will be minimal

gemma_model = "gemma3:270m"
llama_model = "llama3.2:latest"

gemma_system = "You are a very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting."

llama_system = "You are a chatbot who is very argumentative and ragebaiting; \
you disagree with anything in the conversation and you challenge everything, in a snarky way."

gemma_messages = ["Hi there"]
llama_messages = ["Hi"]
````

````python
def call_gemma():
    messages = [{"role": "system", "content": gemma_system}]
    for gemma,llama  in zip(gemma_messages, llama_messages):
        messages.append({"role": "assistant", "content": gemma})
        messages.append({"role": "user", "content": llama})
    response = ollama.chat.completions.create(model=gemma_model, messages=messages)
    return response.choices[0].message.content
````

````python
call_gemma()
````

````python
def call_llama():
    messages = [{"role": "system", "content": llama_system}]
    for gemma, llama in zip(gemma_messages, llama_messages):
        messages.append({"role": "user", "content": gemma})
        messages.append({"role": "assistant", "content": llama})
    messages.append({"role": "user", "content": gemma_messages[-1]})
    response = ollama.chat.completions.create(model=llama_model, messages=messages)
    return response.choices[0].message.content
````

````python
call_llama()
````

````python
call_gemma()
````

````python
gemma_messages = ["Hi there"]
llama_messages = ["Hi"]

display(Markdown(f"### Gemma:\n{gemma_messages[0]}\n"))
display(Markdown(f"### llama:\n{llama_messages[0]}\n"))

for i in range(5):
    gpt_next = call_gemma()
    display(Markdown(f"### Gemma:\n{gpt_next}\n"))
    gemma_messages.append(gpt_next)
    
    claude_next = call_llama()
    display(Markdown(f"### llama:\n{claude_next}\n"))
    llama_messages.append(claude_next)
````
