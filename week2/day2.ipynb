```python
# Connect to OpenAI, Anthropic and Google; comment out the Claude or Google lines if you're not using them

anthropic_url = "https://api.anthropic.com/v1/"
gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
openrouter_url="https://openrouter.ai/api/v1"
ollama_url = "http://localhost:11434/v1"

#anthropic = OpenAI(api_key=anthropic_api_key, base_url=anthropic_url)
ollama = OpenAI(api_key = "ollama",base_url=ollama_url)
openrouter = OpenAI(api_key = openrouter_api_key, base_url=openrouter_url)
gemini = OpenAI(api_key=gemini_api_key, base_url=gemini_url)
```

```python
# Let's wrap a call to ollama in a simple function

system_message = "You are a helpful assistant"

def message_gemini(prompt):
    messages = [{"role": "system", "content": system_message}, {"role": "user", "content": prompt}]
    response = gemini.chat.completions.create(model="gemini-2.5-flash-lite", messages=messages)
    return response.choices[0].message.content
```

```python
# This can reveal the "training cut off", or the most recent date in the training data

message_gemini("What is today's date?")
```

## Forcing dark/light mode

Gradio appears in light mode or dark mode depending on the settings of the browser and computer. There is a way to force gradio to appear in dark mode, but Gradio recommends against this as it should be a user preference (particularly for accessibility reasons). But if you wish to force dark mode for your screens, below is how to do it.

```python
# Define this variable and then pass js=force_dark_mode when creating the Interface 

force_dark_mode = """
function refresh() {
    const url = new URL(window.location);
    if (url.searchParams.get('__theme') !== 'dark') { 
        url.searchParams.set('__theme', 'dark');
        window.location.href = url.href;
    }
}
"""
force_light_mode = """
function refresh() {
    const url = new URL(window.location);
    if (url.searchParams.get('__theme') == 'dark') { 
        url.searchParams.set('__theme', 'light');
        window.location.href = url.href;
    }
}
"""
gr.Interface(fn=shout, inputs="textbox", outputs="textbox", flagging_mode="never", js=force_light_mode).launch(js=force_light_mode)
```

Imagine you have a magical toy robot that has a giant box of toys inside its tummy.

If you ask a normal robot for toys, it will turn the box upside down and dump **ALL** the toys on your head at the exact same time! *CRASH!* That’s a big mess, right? (In computer language, dumping everything at once is called `return`).

But the magic robot is different. Inside this robot is a special, polite button called **`yield`**.

Here is how the `yield` button works:

1. **You ask for a toy.** 2. The robot reaches into its tummy, pulls out just **ONE** toy—like a shiny red racecar—hands it to you, and then... *FREEZES!* 3. It completely pauses and waits. It doesn't drop any more toys. This pause is the **`yield`**.
2. You get to play with the racecar for as long as you want.
3. When you are finally ready for another toy, you tap the robot and say, "Next!"
4. The robot magically *unfreezes*, reaches into its tummy, hands you a bouncy ball, and *freezes* again.

> **In short:** `yield` is a magic pause button! It tells the computer to give you just *one* thing at a time and wait patiently until you are ready for the next one, so you don't get overwhelmed with a giant mess!

When you are learning Python, `yield` is the magic word that turns a normal function into something called a **generator**.

To understand it as a beginner, it helps to compare it directly to the `return` statement you likely already know.

### `return` vs. `yield`

* **`return`** says: "I have done all the work. Here is the final, complete list. I am shutting down now."
* **`yield`** says: "Here is the first piece of the result. I am going to pause right here and remember my exact state until you ask me for the next piece."

---

### A Python Example: The Counter

Let's build a simple function that counts up to a certain number.

Here is what it looks like using **`yield`**:


```python
def count_up_to(max_number):
    count = 1
    while count <= max_number:
        yield count  # The function PAUSES here and hands you the number
        count += 1   # When you ask for the next number, it unpauses here

# 1. We create the generator
my_counter = count_up_to(3)

# 2. We ask for the numbers one at a time using the next() function
print(next(my_counter))  # Output: 1
print(next(my_counter))
print(next(my_counter))


# If we try to call next() again, Python will throw a "StopIteration" error 
# because the generator has finished counting to 3!
```


### Why is `yield` so powerful?

You might be wondering, *"Why not just put all the numbers in a list and `return` the list?"* For small things (like counting to 3), returning a list is perfectly fine. But `yield` shines when things get huge:

* **It saves massive amounts of memory:** If you ask a standard function to return a list of 100 million numbers, Python has to build that giant list and store it all in your computer's memory at once. It might crash your program! A `yield` generator only ever holds **one** number in memory at a time, no matter how high it counts.
* **It is faster to start:** If a function has to process a lot of data, `yield` lets you start working with the very first piece of data immediately, while a standard `return` function makes you wait until every single piece is finished processing.

---

Would you like to see how to smoothly pull items out of a generator using a `for` loop, or would you prefer to see a real-world example of `yield` saving a program from running out of memory?

```python
# Let's create a call that streams back results
# If you'd like a refresher on Generators (the "yield" keyword),
# Please take a look at the Intermediate Python guide in the guides folder

def stream_gemini(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
      ]
    stream = ollama.chat.completions.create(
        model='llama3.2', #model = 'Gemini-2.5-flash'
        messages=messages,
        stream=True
    )
    result = ""
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        yield result
```

```python
message_input = gr.Textbox(label="Your message:", info="Enter a message for GEMINi-FLASH", lines=7)
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_gemini,
    title="GEMINI", 
    inputs=[message_input], 
    outputs=[message_output], 
    examples=[
        "Explain the Transformer architecture to a layperson",
        "Explain the Transformer architecture to an aspiring AI engineer",
        ], 
    flagging_mode="never"
    )
view.launch()
```

```python
def stream_openrouter(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
      ]
    stream = openrouter.chat.completions.create(
        model='openai/gpt-oss-120b:free',
        messages=messages,
        stream=True
    )
    result = ""
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        yield result
```

```python
message_input = gr.Textbox(label="Your message:", info="Enter a message for gpt-oss-120b", lines=7)
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_openrouter,
    title="GPT-OSS", 
    inputs=[message_input], 
    outputs=[message_output], 
    examples=[
        "Explain the Transformer architecture to a layperson",
        "Explain the Transformer architecture to an aspiring AI engineer",
        ], 
    flagging_mode="never"
    )
view.launch()
```

## And now getting fancy

Remember to check the Intermediate Python Guide if you're unsure about generators and "yield"

```python
def stream_model(prompt, model):
    if model=="gemini":
        result = stream_gemini(prompt)
    elif model=="GPT-OSS":
        result = stream_openrouter(prompt)
    else:
        raise ValueError("Unknown model")
    yield from result
```

```python
message_input = gr.Textbox(label="Your message:", info="Enter a message for the LLM", lines=7)
model_selector = gr.Dropdown(["gemini", "GPT-OSS", "ollama"], label="Select model", value="GPT-OSS")
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_model,
    title="LLMs", 
    inputs=[message_input, model_selector], 
    outputs=[message_output], 
    examples=[
            ["Explain the Transformer architecture to a layperson", "GPT-OSS"],
            ["Explain the Transformer architecture to an aspiring AI engineer", "gemini"]
        ], 
    flagging_mode="never"
    )
view.launch()
```

```python
from scraper import fetch_website_contents
```

```python

# Again this is typical Experimental mindset - I'm changing the global variable we used above:

system_message = """
You are an assistant that analyzes the contents of a company website landing page
and creates a short brochure about the company for prospective customers, investors and recruits.
Respond in markdown without code blocks.
"""
```

```python
import scraper

def stream_brochure(company_name, url, model):
    yield ""
    prompt = f"Please generate a company brochure for {company_name}. Here is their landing page:\n"
    prompt += scraper.fetch_website_contents(url)
    if model=="gemini":
        result = stream_gemini(prompt)
    elif model=="GPT-OSS":
        result = stream_openrouter(prompt)
    else:
        raise ValueError("Unknown model")
    yield from result
```

```python

name_input = gr.Textbox(label="Company name:")
url_input = gr.Textbox(label="Landing page URL including http:// or https://")
model_selector = gr.Dropdown(["gemini", "GPT-OSS"], label="Select model", value="GPT-OSS")
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_brochure,
    title="Brochure Generator", 
    inputs=[name_input, url_input, model_selector], 
    outputs=[message_output], 
    examples=[
            ["Hugging Face", "https://huggingface.co", "GPT-OSS"],
            ["Edward Donner", "https://cnn.com", "gemini"]
        ], 
    flagging_mode="never"
    )
view.launch(share=True)
```
